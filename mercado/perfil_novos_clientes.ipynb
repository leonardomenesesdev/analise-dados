## Perfil dos "Novos Clientes"
---
 Analisar as características dos clientes classificados como "Novo Cliente" para entender melhor o perfil de quem a empresa está atraindo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_clientes = pd.read_excel('mercado.xlsx')
df_clientes.head()
```

### Tabela de novos clientes

```python
novos_clientes = df_clientes[df_clientes['Tipo de Cliente'] == 'Novo Cliente']
novos_clientes.head()
```

### Distribuição de Novos Clientes por Centro de Responsabilidade
---
Analisar de quais centros (ou regiões) vêm os novos clientes para entender 
quais localidades estão trazendo mais clientes novos.

```python
centros = novos_clientes['Centro de Responsabilidade'].value_counts()
centros.plot(kind='bar', color='skyblue', edgecolor = 'black')
plt.title("Novos clientes por Centro de Responsabilidade")
plt.xlabel("Centro")
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig('graficos/novos_clientes_por_centro.png')
```

### Distribuição de novos clientes por vendedor
---
Entender o percentual de novos clientes que cada vendedor trouxe

```python
vendedor = novos_clientes['Vendedor'].value_counts()
num_vendedores = len(vendedor)
palette = sns.color_palette("hls", num_vendedores)

vendedor.plot(
    kind="pie",
    autopct='%1.1f%%',
    figsize=(14, 10),
    labels=None,
    colors=palette
)

plt.title("Novos clientes por Vendedor", fontsize=16)
plt.ylabel("")
plt.legend(
    title="Novos clientes x Vendedores",
    labels=vendedor.index,
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

plt.tight_layout()
plt.savefig("graficos/vendedor2.png")
plt.show()
```
