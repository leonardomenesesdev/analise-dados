## Percentual de motivos para contato
---
Entender a distribuição dos motivos para contato dos clientes e identificar se há alguma motivo em destaque

```python
import pandas as pd
import matplotlib.pyplot as plt
df_clientes = pd.read_excel('mercado.xlsx')
```

```python
motivo_contato = df_clientes['Motivo do contato'].value_counts()

motivo_contato.plot(kind = 'pie',
    autopct='%1.1f%%',      # mostra os percentuais
    figsize=(14, 10),       
    labels=None    
)
plt.title('Percentuais para motivos de contato')
plt.ylabel("")
plt.legend(
    title='Motivo contato',
    labels=motivo_contato.index,
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()
plt.savefig('graficos/motivo_contato.png')
```
