```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('orcamento.xlsx')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

### Conta as despesas

```python
# Filtra apenas as transações de 'Despesa'
despesas_df = df[df['Nome conta 1º nível'] == 'Despesa']

# Conta a frequência de TODAS as categorias de 3º nível
contagem_total_categorias = despesas_df['Nome conta 3º nível'].value_counts()

contagem_total_categorias
```

### Gera o gráfico

```python


contagem_total_categorias.plot(
    kind='pie',
    autopct='%1.1f%%',      # mostra os percentuais
    figsize=(14, 10),       
    labels=None             
)

plt.title('Maiores Despesas (Janeiro 22 - Fevereiro 23)', fontsize=16)
plt.ylabel('')

plt.legend(
    title='Categorias',
    labels=contagem_total_categorias.index,
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()



plt.savefig('graficos/frequencia_despesas_pie.png')

print("Gráfico 'frequencia_despesas_pie.png' foi gerado com sucesso.")
```
